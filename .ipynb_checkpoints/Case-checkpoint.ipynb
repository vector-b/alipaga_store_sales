```python
#Apontamos as correlações de pearson em todas as colunas entre si 
corr_df = df.drop(columns=["row_id"])
corr_geral = corr_df.corr(method="pearson")
sns.heatmap(corr_geral, annot=True)
```

Com a tabela acima podemos notar algumas coisas sobre o shipping_cost:
* Há uma relação fortíssima entre sales e shipping_cost, isso se deve ao fato do cliente claramente ter pago o valor de entrega de um produto e por isso impacta diretamente no valor da venda
* Temos uma relação moderada entre profit e shipping_cost, provavelmente também está relacionado ao fato do lucro ser maior conforme o preço de entrega cresce
* Há uma leve relação de crescimento entre shipping_cost e quantidade de itens vendidos, talvez seja um fator que não impacter muito na quantidade de vendas.
    

Podemos destacar algumas coisas sobre o Discount:
* Podemos conferir que a relação discount/profit é muito levemente negativa, o que quer dizer que o lucro de fato é menor conforme o desconto atríbuido, o que faz sentido.
* No geral o fator desconto parece não impactar muito nas outras colunas

```python
outliers=[]
def detect_outlier(data_1):
    threshold=3
    mean_1 = np.mean(data_1)
    std_1 =np.std(data_1)
    for y in data_1:
        z_score= (y - mean_1)/std_1 
        if np.abs(z_score) > threshold:
            outliers.append(y)
    return outliers
```

```python
outlier_datapoints = detect_outlier(dff_q["quantity"])
print(outlier_datapoints[0])
print(dff_q[dff_q["quantity"] == outlier_datapoints[0]])
```

### Clusterização dos Países - Utilizando KMeans

```python
data_ctr = df[["country", "sales","quantity","profit"]]
data_ctr
```

```python
data_ctr = data_ctr.groupby('country')["sales","quantity","profit"].sum().reset_index()
```

```python
data_ctr = data_ctr.sort_values(by="country").drop(columns="country")
```

```python
data_ctr
```

```python
sns.pairplot(data = data_ctr)
```

```python
X_1 = np.array(data_ctr)
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from yellowbrick.cluster import KElbowVisualizer

model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12))
```

```python
visualizer.fit(X_1)
visualizer.show()
```

```python
km = KMeans(n_clusters=3, init='k-means++', n_init=10, max_iter=1000, tol=0.0001, precompute_distances=True, verbose=0, random_state=None, n_jobs=1)
km.fit(X_1)
predicted = km.predict(X_1)
```

```python
data_ctr["cluster"] = predicted
data_ctr["country"] = df["country"]
```

```python
from sklearn.decomposition import PCA
clusters = km.predict(X_1)
pca = PCA(n_components=2)
two_dim = pca.fit_transform(X_1)
scatter_x = two_dim[:, 0] 
scatter_y = two_dim[:, 1]
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')

fig, ax = plt.subplots()
fig.set_size_inches(20,10)

cmap = {0: 'green', 1: 'blue', 2: 'red'}

for group in np.unique(clusters):
    ix = np.where(clusters == group)
    ax.scatter(scatter_x[ix], scatter_y[ix], c=cmap[group], label=group)

ax.legend()
plt.xlabel("PCA 0")
plt.ylabel("PCA 1")
plt.show()
```

```python
data_ctr.sort_values(by="cluster", ascending=False)
```
